- ORDER BY

- ASC -> crescente;
- DESC -> decrescente.

### Exercício 14-a: 

Query que retorne apenas alunos com primeira letra do nome ou do sobrenome sendo vogais;

```python
query = """
SELECT 
    id_aluno,
    cpf AS CPF,
    primeiro_nome AS "Primeiro nome",
    ultimo_nome AS "Último nome",
    telefone_a AS "Primeiro telefone",
    telefone_b AS "Segundo telefone",
    id_cidade AS "Cidade", 
    data_ingresso AS "Data de Ingresso"
FROM 
    alunos
WHERE 
    (primeiro_nome LIKE 'A%' OR primeiro_nome LIKE 'E%' OR primeiro_nome LIKE 'I%' OR primeiro_nome LIKE 'O%' OR primeiro_nome LIKE 'U%')
    OR
    (ultimo_nome LIKE 'A%' OR ultimo_nome LIKE 'E%' OR ultimo_nome LIKE 'I%' OR ultimo_nome LIKE 'O%' OR ultimo_nome LIKE 'U%');
"""



ps.sqldf(query)
```

### Exercício 14-b: 

Query que retorne apenas alunos com primeira letra do nome ou do sobrenome sendo vogais;

```python
query = """
SELECT 
    id_aluno,
    cpf AS CPF,
    primeiro_nome AS "Primeiro nome",
    ultimo_nome AS "Último nome",
    telefone_a AS "Primeiro telefone",
    telefone_b AS "Segundo telefone",
    id_cidade AS "Cidade", 
    data_ingresso AS "Data de Ingresso"
FROM 
    alunos
WHERE 
    (primeiro_nome LIKE 'A%' OR primeiro_nome LIKE 'E%' OR primeiro_nome LIKE 'I%' OR primeiro_nome LIKE 'O%' OR primeiro_nome LIKE 'U%')
    AND
    (ultimo_nome LIKE 'A%' OR ultimo_nome LIKE 'E%' OR ultimo_nome LIKE 'I%' OR ultimo_nome LIKE 'O%' OR ultimo_nome LIKE 'U%');
"""



ps.sqldf(query)
```

### Exercício 15:

Query que retorne apenas Baianos que nomes com mais de 4 letras;

```python
query = """
SELECT 
    a.id_aluno,
    a.primeiro_nome,
    a.ultimo_nome,
    c.cidade,
    c.estado
FROM 
    alunos a
JOIN 
    cidades c ON a.id_cidade = c.id_cidade
WHERE 
    c.estado = 'BA' AND
    LENGTH(a.primeiro_nome) >= 4
LIMIT 10

"""


ps.sqldf(query)
```
